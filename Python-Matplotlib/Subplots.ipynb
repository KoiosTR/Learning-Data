```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilo'nun Yaşa Göre Değişim Grafiği")
```

```python
numpyDizisi1 = np.linspace(0,10,21) # 20 parçaya böldü
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 **3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"y*-")
```

```python
plt.plot(numpyDizisi2,numpyDizisi1,"b--")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2)
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.8,0.8,0.3,0.3])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

```python
myFigure2 = plt.figure()

eksen1 = myFigure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = myFigure2.add_axes([0.3,0.6,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"y")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlığı")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("Küçük Grafik Başlığı")

```

```python
(myFigure,myFigure2) = plt.subplots(nrows=1,ncols=2)
type(myFigure)

for i in myFigure2:
    i.plot(numpyDizisi1,numpyDizisi2,"g")
    i.set_xlabel("X Ekseni")
    i.set_title("Grafik")
    
plt.tight_layout
```
