```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
# Rastgele veri oluşturma
np.random.seed(0)
X = np.random.rand(100, 1) * 10  # 0 ile 10 arasında 100 rastgele sayı
y = 2 + 3 * X + 0.5 * (X ** 2) + np.random.randn(100, 1)  # Polinom fonksiyonu + gürültü
```

```python
# Polinom özellikleri oluşturma
degree = 2  # Polinom derecesi
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X)
```

```python
# Modeli oluşturma ve eğitme
model = LinearRegression()
model.fit(X_poly, y)
```

```python
# Tahmin yapma
X_new = np.linspace(0, 10, 100).reshape(-1, 1)
X_new_poly = poly_features.transform(X_new)
y_new = model.predict(X_new_poly)
```

```python
# Sonuçları görselleştirme
plt.scatter(X, y, color='blue', label='Veri Noktaları')
plt.plot(X_new, y_new, color='red', label='Polinom Regresyon Modeli')
plt.title('Polinom Regresyon Örneği')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
