```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV,cross_val_score
from sklearn.metrics import mean_squared_error,r2_score

from sklearn import model_selection
from sklearn.neighbors import KNeighborsRegressor

from warnings import filterwarnings
filterwarnings('ignore')
```

# Train-Test

```python
df = pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
x = df.drop(["fiyat"] , axis=1)
y = df["fiyat"]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Model Oluşturma

```python
knn_model = KNeighborsRegressor()
```

```python
knn_model
```

```python
model = knn_model.fit(x_train,y_train)
```

```python
model
```

```python
new_data = [[2],[1],[1],[1],[1],[2]]
```

```python
new_data = pd.DataFrame(new_data).T
```

```python
new_data
```

```python
model.predict(new_data)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred[0:10]
```

# Test Hatası Öğrenme

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

# Eğitim Hatası

```python
np.sqrt(mean_squared_error(y_train,model.predict(x_train)))
```

```python
for k in range(30) :
    k=k+1
    k_model = KNeighborsRegressor(n_neighbors = k).fit(x_train,y_train)
    y_pred = k_model.predict(x_train)
    hata=np.sqrt(mean_squared_error(y_train,y_pred))
    
    print("k değeri = ",k, " hata değeri = ",hata)
```

# en ideal k değeri bulma

```python
knn_values = {"n_neighbors": np.arange(1,30,1)}
```

```python
knn_cv_model = GridSearchCV(knn,knn_values,cv=10)
```
