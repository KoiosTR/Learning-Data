```python
sns.countplot(x="type", data=netflix)
fig = plt.gcf()
fig.set_size_inches(5,5)
plt.title("tür")
```

```python
sns.countplot(netflix["rating"])
sns.countplot(netflix["rating"]).set_xticklabels(sns.countplot(netflix["rating"]).get_xticklabels(),rotation=90,ha="right")
fig = plt.gcf()
fig.set_size_inches(8,8)
plt.title("rating")
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x="rating",hue="type", data=netflix)
plt.title("type and rating rate")
plt.show()
```

```python
labels=["Movie","TV Show"]
size=netflix["type"].value_counts()
colors=plt.cm.Wistia(np.linspace(0,1,2))
explode=[0,0.1,0]
plt.rcParams["figure.figsize"]=(9,9)
plt.pie(size,explode=explode,labels=labels,colors=colors,shadow=True,startangle=90) 
plt.title("Type Shows",fontsize=25)
plt.legend()
plt.show()
```
